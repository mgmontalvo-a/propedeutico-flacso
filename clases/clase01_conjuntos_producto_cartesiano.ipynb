## 3. Conjuntos: de la pizarra a Python

En matemáticas, un **conjunto** es una colección **no ordenada** de elementos **distintos**:

$$A = \{1, 2, 3, 4\}$$

Decimos que $2 \in A$ («2 pertenece a $A$»), que $7 \notin A$, y llamamos **cardinalidad** $|A|$ al número de elementos.

En Python la traducción es casi literal:

| Matemáticas | Python | Se lee |
|---|---|---|
| $A = \{1, 2, 3\}$ | `A = {1, 2, 3}` | el conjunto $A$ |
| $2 \in A$ | `2 in A` | 2 pertenece a $A$ |
| $7 \notin A$ | `7 not in A` | 7 no pertenece a $A$ |
| $\lvert A \rvert$ | `len(A)` | cardinalidad de $A$ |
| $\emptyset$ | `set()` | conjunto vacío |

> ⚠️ **Trampa clásica:** `{}` **no** crea un conjunto vacío sino un *diccionario* (otra estructura que veremos más adelante). El conjunto vacío se escribe `set()`.

### 🧑‍🏫 Yo hago 3.1 — Definir un conjunto, pertenencia y cardinalidad

````python
# Un conjunto se define con llaves
A = {1, 2, 3, 4}
print("A =", A)
print("Cardinalidad |A| =", len(A))

# Pertenencia: el operador 'in' devuelve True (verdadero) o False (falso)
print("¿2 ∈ A?", 2 in A)
print("¿7 ∈ A?", 7 in A)
print("¿7 ∉ A?", 7 not in A)
````

### 🧑‍🏫 Yo hago 3.2 — Dos propiedades que Python respeta solo

Los conjuntos **eliminan duplicados automáticamente** y **no tienen orden**. Ambas propiedades son matemáticas, no caprichos del lenguaje: $\{1, 2, 2\} = \{1, 2\}$ y $\{1, 2\} = \{2, 1\}$.

````python
# Propiedad 1: los duplicados se eliminan solos
votos = {"Ana", "Luis", "Ana", "María", "Luis"}
print("Personas distintas que votaron:", votos)
print("Número de votantes:", len(votos))

# Propiedad 2: un conjunto NO tiene orden interno.
# El orden en que se imprime puede variar entre sesiones; si necesitas
# una presentación ordenada, usa sorted(), que devuelve una lista ordenada.
print("Presentación ordenada:", sorted(votos))

# El conjunto vacío se crea con set(), no con {}
vacio = set()
print("Conjunto vacío:", vacio, "| cardinalidad:", len(vacio))
````

## 4. Operaciones entre conjuntos

Las cuatro operaciones básicas y la relación de contención, con su traducción a Python:

| Operación | Definición | Python |
|---|---|---|
| Unión | $A \cup B = \{x : x \in A \ \text{o} \ x \in B\}$ | `A \| B` |
| Intersección | $A \cap B = \{x : x \in A \ \text{y} \ x \in B\}$ | `A & B` |
| Diferencia | $A \setminus B = \{x : x \in A \ \text{y} \ x \notin B\}$ | `A - B` |
| Diferencia simétrica | $A \triangle B = (A \setminus B) \cup (B \setminus A)$ | `A ^ B` |
| Subconjunto | $A \subseteq B$ | `A <= B` |

> 💡 También existen los métodos equivalentes `A.union(B)`, `A.intersection(B)`, `A.difference(B)`, `A.symmetric_difference(B)` y `A.issubset(B)`. Usaremos los operadores porque se parecen más a la notación matemática.

### 🧑‍🏫 Yo hago 4.1 — Las cuatro operaciones con números

````python
A = {1, 2, 3, 4, 5}
B = {4, 5, 6, 7}

print("A ∪ B =", A | B)     # unión: elementos en A o en B (o en ambos)
print("A ∩ B =", A & B)     # intersección: elementos en A y en B a la vez
print("A \\ B =", A - B)     # diferencia: en A pero no en B
print("B \\ A =", B - A)     # ¡la diferencia NO es conmutativa!
print("A △ B =", A ^ B)     # diferencia simétrica: en uno u otro, pero no en ambos

# Contención
print("¿{1, 2} ⊆ A?", {1, 2} <= A)
print("¿A ⊆ B?", A <= B)
````

### 🧑‍🏫 Yo hago 4.2 — Ejemplo aplicado: geografía política de Sudamérica

Sea el conjunto de países **andinos** (atravesados por la cordillera) y el de países **amazónicos** (miembros de la OTCA, la Organización del Tratado de Cooperación Amazónica). Las operaciones de conjuntos responden preguntas sustantivas de forma inmediata.

````python
andinos = {"Venezuela", "Colombia", "Ecuador", "Perú", "Bolivia", "Chile", "Argentina"}
amazonicos = {"Brasil", "Bolivia", "Colombia", "Ecuador", "Guyana", "Perú", "Surinam", "Venezuela"}

print("Andinos y amazónicos a la vez:", sorted(andinos & amazonicos))
print("Andinos que no son amazónicos:", sorted(andinos - amazonicos))
print("Amazónicos que no son andinos:", sorted(amazonicos - andinos))
print("Total de países mencionados:", len(andinos | amazonicos))
````

## 5. Producto cartesiano

El **producto cartesiano** de $A$ y $B$ es el conjunto de todos los **pares ordenados** cuya primera componente está en $A$ y la segunda en $B$:

$$A \times B = \{(a, b) : a \in A \ \text{y} \ b \in B\}$$

Dos ideas clave:

- En un **par ordenado el orden importa**: $(1, 2) \neq (2, 1)$. En Python, un par ordenado es una **tupla**: `(a, b)`.
- **Regla del producto:** $|A \times B| = |A| \cdot |B|$. Este principio de conteo es la base de los diseños muestrales y experimentales que verás en estadística.

El nombre no es casual: el **plano cartesiano** de Descartes es exactamente $\mathbb{R} \times \mathbb{R} = \mathbb{R}^2$, el conjunto de todos los pares ordenados de números reales.

### 🧑‍🏫 Yo hago 5.1 — Tuplas: donde el orden sí importa

````python
# Una tupla es una pareja (o n-tupla) ORDENADA; se escribe con paréntesis
par1 = (1, 2)
par2 = (2, 1)

print("par1 =", par1)
print("par2 =", par2)
print("¿(1, 2) == (2, 1)?", par1 == par2)      # False: en las tuplas el orden importa

# Compara con los conjuntos, donde el orden NO importa
print("¿{1, 2} == {2, 1}?", {1, 2} == {2, 1})  # True
````

### 🧑‍🏫 Yo hago 5.2 — Construir $A \times B$ por comprensión

Python permite escribir conjuntos **por comprensión**, calcando la notación matemática. Compara:

$$A \times B = \{(a, b) : a \in A, \ b \in B\}$$

```python
AxB = {(a, b) for a in A for b in B}
```

Se lee igual: «el conjunto de las parejas $(a, b)$, para cada $a$ en $A$ y cada $b$ en $B$».

````python
A = {1, 2, 3}
B = {"x", "y"}

# Producto cartesiano por comprensión
AxB = {(a, b) for a in A for b in B}

print("A × B =", sorted(AxB))
print("|A| =", len(A), " |B| =", len(B), " |A × B| =", len(AxB))
print("¿Se cumple |A × B| == |A| · |B|?", len(AxB) == len(A) * len(B))
````

### 🧑‍🏫 Yo hago 5.3 — La herramienta estándar y la no conmutatividad

Para productos cartesianos, la biblioteca estándar de Python trae `itertools.product`. La usamos para comprobar que el producto cartesiano **no es conmutativo**: $A \times B \neq B \times A$ (los pares están invertidos).

````python
from itertools import product   # herramienta estándar para productos cartesianos

BxA = set(product(B, A))
print("B × A =", sorted(BxA))
print("¿A × B == B × A?", AxB == BxA)   # False: el producto cartesiano no es conmutativo
````

### 🧑‍🏫 Yo hago 5.4 — Ver el producto cartesiano

Si $A$ y $B$ son conjuntos de números, $A \times B$ forma una **malla de puntos** en el plano: la intuición geométrica del producto cartesiano. Graficamos con `matplotlib`, la biblioteca de visualización más usada en Python (ya viene instalada en Colab).

````python
import matplotlib.pyplot as plt

A = {1, 2, 3}
B = {1, 2}
AxB = {(a, b) for a in A for b in B}

xs = [par[0] for par in AxB]   # primeras coordenadas
ys = [par[1] for par in AxB]   # segundas coordenadas

plt.figure(figsize=(5, 3.2))
plt.scatter(xs, ys, s=80, color="darkblue")
plt.title("A × B con A = {1, 2, 3} y B = {1, 2}")
plt.xlabel("A (primera coordenada)")
plt.ylabel("B (segunda coordenada)")
plt.xticks([1, 2, 3])
plt.yticks([1, 2])
plt.grid(True, linestyle=":")
plt.show()
````
